#Sigmoid

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

data = np.linspace(-10,10)
# print(len(data))
plt.plot(data,sigmoid(data))
plt.axis('tight')
plt.title('Activation Function : SIGMOID')
plt.show()
```

#Tanh

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tanh(x):
    return np.tanh(x)

data = np.linspace(-10,10)

plt.plot(data,tanh(data))
plt.axis('tight')
plt.title('Activation Function : TANH')
plt.show()
```

#RELU

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def RELU(x):
    return np.maximum(0,x)
data = np.linspace(-10,10)
plt.plot(data,RELU(data))
plt.axis('tight')
plt.title('Activation Function : RELU')
plt.show()
```

#Softmax

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def softmax(x):
    return np.exp(x)/np.exp(x).sum()
data = np.linspace(-10,10)
plt.plot(data,softmax(data))
plt.axis('tight')
plt.title('Activation Function : SOFTMAX')
plt.show()
```

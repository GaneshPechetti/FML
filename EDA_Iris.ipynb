```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

#current version of seaborn generates errors that will be ignored
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline 
#to draw plots immediately after cell
```

```python
iris = pd.read_csv("IRIS.csv")
print(iris.head())
print(iris.shape)
```

```python
iris.describe()
```

```python
iris.species.unique()
```

```python
iris['species'].value_counts()
```

```python
sns.relplot(iris["sepal_length"],iris["sepal_width"],hue=iris["species"])
plt.show()
```

```python
sns.pairplot(data=iris,hue="species")
plt.show()
```

```python
sns.heatmap(iris.corr(),annot=True,fmt="f")
plt.show()
```

```python
sns.boxplot("species","petal_length",data=iris)
plt.show()
```

```python
sns.violinplot("species","petal_length",data=iris)
plt.show()
```

```python
iris.hist(edgecolor="black",bins=25,figsize=(12,6))
plt.show()
```

```python
#Filtering by species
indices = iris['species'] == "Iris-setosa"
setosa = iris.loc[indices,:]
indices = iris['species'] == "Iris-versicolor"
versicolor = iris.loc[indices,:]
indices = iris['species'] == "Iris-virginica"
virginica = iris.loc[indices,:]

print(setosa.head())
print(versicolor.head())
```

```python
del setosa['species'], versicolor['species'], virginica['species']
```

```python
print(setosa.head())
```

```python
setosa.plot(kind='hist',bins=50,range=(0,8),alpha=0.4)
plt.title("Setosa Dataset")
plt.xlabel('cm')

versicolor.plot(kind='hist',bins=50,range=(0,8),alpha=0.4)
plt.title("Versicolor Dataset")
plt.xlabel('cm')

virginica.plot(kind='hist',bins =50,range=(0,8),alpha=0.4)
plt.title("Virginica Dataset")
plt.xlabel('cm')

plt.show()
```
